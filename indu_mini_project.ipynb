```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_csv("C:/Users/bhava/OneDrive/Pictures/Desktop/GPTC/5th SEM/EXCEL/Mall_Customers.csv")
event_a_column = 'Gender'
event_b_column = 'Age'
conditional_probabilities = data.groupby(event_a_column)[event_b_column].mean()
plt.figure(figsize=(8, 6))  # Set the figure size
sns.barplot(x=conditional_probabilities.index, y=conditional_probabilities.values)
plt.xlabel(event_a_column)
plt.ylabel(f'P({event_b_column} | {event_a_column})')
plt.title('Conditional Probability Chart Example')
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility if needed
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_csv("C:/Users/bhava/OneDrive/Pictures/Desktop/GPTC/5th SEM/EXCEL/Mall_Customers.csv")
data_column = 'Age'
plt.figure(figsize=(8, 6))  # Set the figure size
sns.kdeplot(data=data[data_column], shade=True, color='red')
plt.xlabel(data_column)
plt.ylabel('Density')
plt.title('Density Plot Example')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv("C:/Users/bhava/OneDrive/Pictures/Desktop/GPTC/5th SEM/EXCEL/Mall_Customers.csv")
categories = data['Gender']
values = data['Age']
plt.figure(figsize=(9, 4))  # Set the figure size
plt.pie(values, labels=categories, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart Example')  # Add a title
plt.axis('equal')  # Equal aspect ratio ensures that the pie is drawn as a circle.
plt.show()
```
